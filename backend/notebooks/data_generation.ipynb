```python
import pandas as pd
import random
import uuid
import os

# 1. Define the Indian Medical Landscape
specialties = {
    "Cardiology": ["Interventional Cardiologist", "Pediatric Cardiologist", "Electrophysiologist"],
    "Neurology": ["Neurosurgeon", "Neurologist", "Stroke Specialist"],
    "Oncology": ["Medical Oncologist", "Surgical Oncologist", "Radiation Oncologist"],
    "Pediatrics": ["General Pediatrician", "Neonatologist", "Pediatric Nephrologist"],
    "Internal Medicine": ["Diabetologist", "Endocrinologist", "Infectious Disease Specialist"],
    "Orthopedics": ["Spine Surgeon", "Sports Medicine Specialist", "Joint Replacement Surgeon"],
    "Psychiatry": ["Child Psychiatrist", "Addiction Specialist", "General Psychiatrist"],
    "Gastroenterology": ["Hepatologist", "Surgical Gastroenterologist", "Medical Gastrologist"],
    "Dermatology": ["Cosmetic Dermatologist", "Trichologist", "Medical Dermatologist"],
    "ENT": ["Otologist", "Rhinologist", "Laryngologist"]
}

cities = ["Mumbai", "Delhi NCR", "Bengaluru", "Hyderabad", "Chennai", "Kolkata", "Pune", "Ahmedabad", "Lucknow", "Chandigarh"]
hospital_chains = ["Apollo Hospital", "Max Super Speciality", "Fortis Healthcare", "Manipal Hospital", "AIIMS", "Medanta", "Narayana Health", "Cloudnine"]

first_names = ["Arjun", "Aditi", "Rahul", "Priya", "Vikram", "Sneha", "Amit", "Anjali", "Suresh", "Meera", "Rohan", "Sanya", "Ishaan", "Kavita"]
last_names = ["Sharma", "Gupta", "Iyer", "Reddy", "Patel", "Verma", "Nair", "Mishra", "Deshmukh", "Choudhury", "Joshi", "Aggarwal"]

# 2. Generation Loop for 2,500+ Doctors
master_data = []
for i in range(1, 2600):  # Slightly over 2500 for good measure
    parent_spec = random.choice(list(specialties.keys()))
    sub_spec = random.choice(specialties[parent_spec])
    city = random.choice(cities)
    
    # Randomly generate a 10-digit Indian phone number
    phone = f"+91-{random.randint(7, 9)}{random.randint(10, 99)}-{random.randint(100, 999)}-{random.randint(1000, 9999)}"
    
    days = random.sample(["Mon", "Tue", "Wed", "Thu", "Fri", "Sat"], k=random.randint(3, 5))
    schedule = f"{', '.join(days)} | {random.randint(9, 10)}:00 AM - {random.randint(4, 7)} PM"

    master_data.append({
        "doctor_uuid": str(uuid.uuid4()),
        "name": f"Dr. {random.choice(first_names)} {random.choice(last_names)}",
        "specialty": parent_spec,
        "sub_specialty": sub_spec,
        "phone": phone,
        "hospital": f"{random.choice(hospital_chains)} - {city}",
        "city": city,
        "experience_years": random.randint(5, 40),
        "rating": round(random.uniform(3.8, 5.0), 1),
        "reviews": random.randint(20, 1500),
        "availability": random.choice(["Immediate", "Tomorrow", "Within 3 Days", "On-Call"]),
        "fees_inr": random.choice([800, 1000, 1200, 1500, 2500, 3000]),
        "schedule": schedule
    })

# 3. Create Directory and Save
# In a Notebook, we use '../' to go from /notebooks up to the root /data
os.makedirs('../data/raw', exist_ok=True)
df = pd.DataFrame(master_data)
df.to_csv('../data/raw/indian_healthcare_master.csv', index=False)

print(f"✅ Successfully generated {len(df)} records!")
print(df.head()) # Preview the data
```
